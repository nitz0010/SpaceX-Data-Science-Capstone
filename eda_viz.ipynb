```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
df = pd.read_csv('dataset_part_1.csv')

# 2. Fix Column Names (removes spaces and converts to lowercase)
df.columns = df.columns.str.strip().str.lower()

# 3. Success Rate by Rocket Type (Replacing 'orbit' with 'rocket')
# Since 'orbit' doesn't exist in your CSV, we use 'rocket'
rocket_success = df.groupby('rocket')['class'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='rocket', y='class', data=rocket_success)
plt.title('Success Rate by Rocket Type')
plt.ylabel('Success Rate')
plt.show()

# 4. Payload vs Launch Site Scatterplot
# In your CSV, the column is 'launchpad' (not 'launchsite')
plt.figure(figsize=(12, 6))
sns.scatterplot(y="launchpad", x="payloadmass", hue="class", data=df)
plt.title('Payload Mass vs Launch Pad Success/Failure')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
