```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# --- STEP 1: LOAD THE DATA (Missing in your previous run) ---
# This URL is the standard dataset for this SpaceX/Falcon9 lab
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
data = pd.read_csv(URL)

# Usually, 'X' comes from a pre-processed file or 'transform' variable 
# If 'transform' is not defined, we use the numeric features:
X = data.select_dtypes(include=[np.number]).drop(['Class'], axis=1)
Y = data['Class'].to_numpy()

# --- STEP 2: SPLIT THE DATA ---
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# --- STEP 3: TRAIN THE MODEL ---
tree = DecisionTreeClassifier()
tree.fit(X_train, Y_train)

# --- STEP 4: PREDICT AND VISUALIZE ---
yhat = tree.predict(X_test)

cm = confusion_matrix(Y_test, yhat)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Decision Tree Confusion Matrix')
plt.show()
```
